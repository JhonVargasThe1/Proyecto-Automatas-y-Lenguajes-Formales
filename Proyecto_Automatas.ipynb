```python
from graphviz import Digraph
def generate_graph(dtm):
    graph = Digraph(comment='Turing Machine')

    for state in dtm.states:
        graph.node(state)

    for state, transitions in dtm.transitions.items():
        for symbol, action in transitions.items():
            next_state, write_symbol, direction = action
            label = f"{symbol}/{write_symbol},{direction}"
            graph.edge(state, next_state, label=label)

    return graph
graph = generate_graph(traductor)
graph.render('turing_machine_graph', format='pdf', cleanup=True)
```
